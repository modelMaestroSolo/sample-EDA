#### Some column transformations

```python
violation_time_df  = df1.dropna(subset=['Violation Time']).copy() #avoid SettingwithCopyWarning
print(violation_time_df.shape)
violation_time_df.isnull().sum()
```

```python
# to help understand the distribution of violations over time in a day
def day(time_str):
    if time_str.endswith('A'):
        return 'BEFORE MIDDAY'
    else:
        return 'AFTER MIDDAY'
```

```python
#dropna to match index
violation_time_df['Time of Day'] = df1.loc[:,'Violation Time'].dropna().apply(day) 
```

```python
violation_time_df['Violation Time'].describe()
```

```python
#anomaly: 0050AM is 1250AM on 12-hour format
violation_time_df[violation_time_df['Violation Time'].str.startswith('00')  ].head()  
```

```python
violation_time_df[ (violation_time_df['Violation Time'].str.startswith('00')) \
                 & (violation_time_df['Violation Time'].str.endswith('P')) ]  
```

```python
# to help understand the distribution of violations over periods in a day
def convert_to_24hour(time_str):
    time_str = time_str + 'M'  # required for string format specified
    if time_str.startswith('00'):
        time_str = '12' + time_str[2:]
    time_obj = time.strptime(time_str, '%I%M%p')
    return time.strftime('%H%M', time_obj)
```

```python
violation_time_df['24-hour Scale'] = df1.loc[:,'Violation Time'].dropna().\
                                                                    apply(convert_to_24hour)
violation_time_df.head()
```

```python
print(df1.shape)
print(violation_time_df.shape)
```

```python
violation_time_df.info()
```

```python
# convert time to numeric to help bin values into specific periods
violation_time_df['24-hour Scale'] = pd.to_numeric(violation_time_df['24-hour Scale'])
```

#### Convert nominal categorical variables into categorical dtype
Done in order to make it easier to work with and analyze the data. 
<br>

categorical dtype are data types specifically designed to handle categorical variables. They enable you to easily perform various operations such as grouping, aggregating, and sorting the data based on the categories.
<br>

When you transform your nominal categorical variables into factors or categorical dtype, it also helps to reduce the memory usage and improve the performance of your code. This is because factors and categorical dtype use integer codes to represent the categories, rather than storing the full string values for each observation.
<br>

For more information on pandas categorical data, please refer to the [Pandas documentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/categorical.html).



```python
violation_time_df[ violation_time_df.select_dtypes(include=['object']).columns ] =\
    violation_time_df.select_dtypes(include=['object']).astype('category')

violation_time_df.dtypes
```

```python
print([a for a in violation_time_df['24-hour Scale'].values])
```

```python
violation_time_df['24-hour Scale'].describe()
```

```python
def period(day):
    if (day >=500) & (day < 1200):
        return 'MORNING'
    elif (day >= 1200) & (day < 1700):
        return 'AFTERNOON'
    elif (day >= 1700) & (day < 2100):
        return 'EVENING'
    else:
        return 'NIGHT'
```

```python
violation_time_df['Period of Day'] = violation_time_df['24-hour Scale'].apply(period)
violation_time_df.head()
```

```python
#### Crosstabulation of frequent offenders and their registration states #############
RegState_frequent_offenders = violation_time_df.set_index('Plate ID').\
                                                loc[frequent_offenders]['Registration State']
RegState_frequent_offenders=RegState_frequent_offenders.\
                    value_counts()[['NY', 'MN', 'IN', 'MA', 'AL', 'NJ']]

RegState_frequent_offenders
```

```python
#freqent offenders registration states
RegState_offenders = violation_time_df['Registration State'].\
                                    value_counts().loc[RegState_frequent_offenders.index]
RegState_df = pd.DataFrame([RegState_frequent_offenders, RegState_offenders], \
                     index=['Frequent Offender', 'Offenders']).transpose()
```

```python
RegState_df.rename_axis('State', inplace=True)
```

```python
RegState_df['Frequent Offender Proportion'] = \
                        RegState_df['Frequent Offender'] / RegState_df['Offenders']
RegState_df['Non-Frequent Offender Proportion'] =\
                        1 - RegState_df['Frequent Offender Proportion']
RegState_df.reset_index(inplace=True)
RegState_df
```

```python
# create stacked bar chart
ax = RegState_df.plot(kind='bar', x='State', y=['Frequent Offender Proportion', 'Non-Frequent Offender Proportion'], stacked=True)

ax.set_ylabel('Proportion')
ax.legend(['Frequent Offender', 'Non-Frequent Offender'], loc='best')

plt.show()
```

### 4.  Conclusions and Recommendations
#### recommendations, limitations, areas for further research<br>

Conclusions:
The exploratory data analysis has revealed important insights into violation trends, popular locations and periods of the day, and offender segmentation. The majority of offenders violated only once, which is a positive indication for road safety. However, there is still a need to address the 4% of frequent offenders who violate regulations multiple times. Broadway street during the mornings and afternoons was identified as a high-risk area, accounting for a significant proportion of all violations. The registration states with the most frequent offenders were identified, providing an opportunity for targeted enforcement efforts.<br>
<br>

Recommendations:
Based on the findings, several recommendations can be made to improve road safety:

1. Develop targeted interventions for high-risk areas such as Broadway street during the mornings and afternoons. This could include the installation of speed cameras, increased police patrols, and public education campaigns.
2. Focus enforcement efforts on vehicles registered in the states with the most frequent offenders, such as 'IN' and 'AL', to deter repeat violations.
3. Implement programs to educate and encourage motorists to comply with regulations, particularly those who have violated once.
4. Continue to monitor violation trends and patterns to inform ongoing efforts to improve road safety.
5. Consider implementing technology-based solutions such as automatic license plate recognition (ALPR) to improve the efficiency of enforcement efforts.
By implementing these recommendations, law enforcement agencies can work towards reducing the number of violations, improving road safety, and creating a safer environment for all road users.<br>
